```python
with open("sgb-words.txt", "r") as f:
    words = [w for w in f.read().split("\n") if len(w) == 5]
print(len(words))
print(words[:10])
```

```python
counts = {}
for word in words:
    for char in word:
        if char not in counts:
            counts[char] = 0
        counts[char] += 1
        
print(sorted(counts.items(), key=lambda x: x[1], reverse=True))
```

```python
singles = {i:{} for i in range(5)}
for word in words:
    for i in range(5):
        char = word[i]
        if char not in singles[i]:
            singles[i][char] = 0
        singles[i][char] += 1
for i in singles:
    print(sorted(singles[i].items(), key=lambda x: x[1], reverse=True), end="\n\n")
```

```python
doubles = {i:{} for i in range(5)}
for word in words:
    for i in range(4):
        char = word[i:i+2]
        if char not in doubles[i]:
            doubles[i][char] = 0
        doubles[i][char] += 1
for i in doubles:
    print(sorted(filter(lambda x: (x[1] != 1), doubles[i].items()), key=lambda x: x[1], reverse=True), end="\n\n")
```
